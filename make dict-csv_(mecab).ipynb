```python
import pandas as pd
from pandas import DataFrame

dataset = pd.read_csv('/Users/SAEROM/reple_result.csv' , encoding='cp949')
#첫 5행을 불러옴,
dataset.head() 
```

```python
data_X = dataset["reple_result"]
#print(data_X)
```

```python
#slang_dict의 형태
#빈도수를 체크해야함..

words_list = []

for data in data_X:
    words = data.split(" ")
    for word in words:
        words_list.append(word)
print(words_list)
```

```python
def words_sorted_by_occurence(words_list):
    d = dict()
    
    for word in words_list:
        if word in d:
            d[word]= d[word]+1
        else :
            d[word]=1
            
    return d
```

```python
slang_dict = dict()
slang_dict = words_sorted_by_occurence(words_list)
sorted(slang_dict.items(), key=lambda x:x[1], reverse=True)
```

```python
import csv
with open('/Users/SAEROM/slang_dict.csv', 'w', encoding='cp949', newline='') as writer_csv:
    field_name_list = ['word', 'frequency']
    writer = csv.DictWriter(writer_csv, fieldnames=field_name_list)
    writer.writeheader()
    for key, value in slang_dict.items():
        writer.writerow({'word':key, 'frequency':value})
```
