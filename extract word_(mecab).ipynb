```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from konlpy.tag import Mecab
import re

mecab = Mecab()
```

```python
from konlpy.tag import Twitter
#norm은 현대적인말 그래욬ㅋㅋㅋ 같은 것을 그래요로 바꾸어 주는 것이다. 
#stem은 그래요를 그렇다처럼 원형으로 바꾸어 주는 것이다.
twitter = Twitter()
```

```python
dataset = pd.read_csv("/Users/SAEROM/reple.csv")
#첫 5행을 불러옴
dataset.head() 
```

```python
data_X = dataset["reple"]
#print(data_X)
```

```python
new_texts = []
for i in range(300):
    pos_ = twitter.pos(data_X[i], norm=True, stem=True)
    sentence = ""
    for pos in pos_:
        if pos[1] in ['Noun', 'KoreanParticle']:
            sentence += pos[0]
            sentence += " "
    sentence = re.sub('ㅋ', '', sentence)
    sentence = re.sub('ㅎ', '', sentence)
    sentence = re.sub('ㅜ', '', sentence)
    sentence = re.sub('ㅠ', '', sentence)
    sentence = re.sub('ㅇㅇ', '', sentence)
    new_texts.append(sentence)
print(new_texts)
```

```python
total_list = []
for repeat in range (300):
    #원본 문자열 
    str_ = mecab.pos(data_X[repeat])
    #print(str_)

    #작업1. 포스태깅 중 필요한 정보만 뽑아 단어 리스트를 만든다.
    #list_pos = []
    str_pos=""
    #'XSV+E', 'XSA+E', 'vv+ec', 'sn' ,'MM', 'XR', 'MAG'
    tag = ['NNG','NNBC', 'NR','MM', 'XSV+EC', 'XSA+EF', 'UNKNOWN', 'IC']
    for s in range(len(str_)): #0번째는 단어가, 1번째는 품사가 결과로 나옴.
        for i in tag:
            if (i in str_[s][1] and str_[s][1] != 'XSN'):
                str_pos += str_[s][0]
            #elif (str_[s][1]=='VV' and str_[s+1][1]=='EC'):
            #    str_pos += str_[s][0]
            #    str_pos += str_[s+1][0]
                #list_pos.append(s[0])
            str_pos = re.sub('ㅋ', '', str_pos)
            str_pos = re.sub('ㅎ', '', str_pos)
            str_pos = re.sub('ㅜ', '', str_pos)
            str_pos = re.sub('ㅠ', '', str_pos)
            str_pos = re.sub('ㅇㅇㅇ', '', str_pos)
        #print("포스태깅 한 리스트 : " , list_pos)
        #print("포스태깅 한 문자열 : " , str_pos)

    #작업2. 문자열을 한 글자씩 분리해 한글자 리스트를 만든다.
    example = data_X[repeat]

    list_one = []
    for i in range(len(example)):
        #print(example[i])
        list_one.append(example[i])
    #print("한글자 리스트 : " , list_one)

    #작업3. 띄어쓰기까지 포함된 새로운 리스트
    str_new = ''
    num = 0

    #공백을 구분하기 위해 새로운 숫자변수 부여
    b_num=0
    for i in range(len(list_one)) : 
        if num < len(str_pos) :
            if list_one[i]==str_pos[num]:
                if i < len(list_one)-3 and list_one[i+2] == ' ' and list_one[i] == list_one[i+3]:
                    continue
                else:
                    str_new += list_one[i]
                    num = num+1
                    b_num = b_num+1
 
            elif list_one[i]==' ' or list_one[i]=='\xa0' or list_one[i] == '.':
                if(b_num == 0):
                    continue
                elif(str_new[b_num-1]==' '):
                    continue
                else :
                    str_new += ' '
                    b_num = b_num +1
                #print(str_new , num)
            else : continue
                #print(str_new , num)
    
    #print(str_new)
    str_new = str_new.strip()
    total_list.append(str_new)
print(total_list)
        
```

```python
data = pd.DataFrame(total_list) 
data.columns = ['reple_result']
data.to_csv('/Users/SAEROM/reple_result.csv', encoding='cp949')
data
```
