```python
data_X = dataset["reple"]
#print(data_X)
```

```python
t = "내가 가다가 말함"
print(twitter.pos(t, norm=True, stem=True))
```

```python
new_texts = []
for i in range(300):
    pos_ = twitter.pos(data_X[i], norm=True, stem=True)
    sentence = ""
    for pos in pos_:
        if pos[1] in ['Noun', 'KoreanParticle']:
            sentence += pos[0]
            sentence += " "
    sentence = re.sub('ㅋ', '', sentence)
    sentence = re.sub('ㅎ', '', sentence)
    sentence = re.sub('ㅜ', '', sentence)
    sentence = re.sub('ㅠ', '', sentence)
    sentence = re.sub('ㅇㅇ', '', sentence)
    sentence = sentence.strip()
    new_texts.append(sentence)
print(new_texts)
```

```python
words_list = []

for text in new_texts:
    words = text.split(" ")
    for word in words:
        if word != '':
            words_list.append(word)
```

```python
def words_sorted_by_occurence(words_list):
    d = dict()
    
    for word in words_list:
        if word in d:
            d[word]= d[word]+1
        else :
            d[word]=1
            
    return d
```
